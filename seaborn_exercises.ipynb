3. Load the `swiss` dataset from pydataset and read it's documentation.

```python
swiss = data('swiss')
data('swiss',show_doc=True)
```

Create visualizations to answer the following questions:
- Create an attribute named `is_catholic` that holds a boolean value of whether or not the province is Catholic. (Choose a cutoff point for what constitutes catholic)

```python
# In order to determine what we'll constitue Catholic to be, let's look at the distribution
swiss.Catholic.hist(bins=5)

plt.show()
```

```python
# swiss.Catholic.mean()
```

```python
# Let's make the province to be Catholic if more than 50% of the population is Catholic
swiss['is_catholic'] = swiss.Catholic > 50
```

- Does whether or not a province is Catholic influence fertility?

```python
swiss[['Fertility','Catholic']].corr()
```

```python
sns.relplot(swiss,y='Fertility',x='Catholic',hue='is_catholic')

plt.show()

# There doesn't appear to be a strong correlation between being Catholicism and fertility
# That being said, fertility does seem to be higher among more predominantly Catholic provinces
```

```python
# From the review
# This looks better and shows a better relation
sns.boxplot(swiss, x='is_catholic',y='Fertility');
```

- What measure correlates most strongly with fertility?

```python
swiss.iloc[:,:len(swiss.columns)-1].corr()[:1]
# .iloc selects all index values and leaves off the last column
# Last bracket drops all resulting indexes except first one

# Accordingly, education correlates most strongly with fertility, although it's a negative correlation
```

```python
sns.heatmap(swiss.corr()[:1],cmap='Reds',annot=True)
```

4. Load the `chipotle` dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
import my_functions as mf

url = mf.get_db_url('chipotle')
```

```python
chipotle = pd.read_sql('select * from orders',url)
chipotle.head()
```

```python
# Fix the item_price
chipotle['item_price'] = chipotle.item_price.str.strip('$').astype('float')
chipotle.head()
```

```python
# Filter down to 4 popular items
popular_items = chipotle.groupby('item_name').sum('quantity').nlargest(4,columns='quantity')
popular_items
```

```python
# Turn it into a graph!
sns.barplot(popular_items,y=popular_items.index,x='item_price')

plt.show()
```
