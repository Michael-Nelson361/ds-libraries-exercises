```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

1. Use matplotlib to plot the following equation:
    
    $$
    y = x^2 - x + 2
    $$
    
    You'll need to write the code that generates the x and y points.
    
    Add an anotation for the point `0, 0`, as the origin.

```python
x = pd.Series(range(-50,50))
y = x**2 - x + 2
```

```python
plt.plot(x,y)
plt.annotate('origin',(0,0))
plt.grid()
plt.show;
```

2. Create and label 4 separate charts with the following equations (choose a range for x that makes sense):
    - $y=√x$
    - $y=x^3$
    - $y=2^x$
    - $y=1/(x+1)$
    
    You can use functions from the `math` module to help implement some of the equations above.

```python
# Make the range
x = pd.Series(range(50))

y1 = x**(1/2)
y1_label = 'y=x^(1/2)'

y2 = x**3
y2_label = 'x^3'

y3 = 2**x
y3_label = '2^x'

y4 = 1/(x+1)
y4_label = '1/(x+1)'
```

```python
fig,ax = plt.subplots(2,2)

ax[0,0].plot(x,y1)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)

plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.plot(x,y1,label=y1_label)
plt.plot(x,y2,label=y2_label)
plt.plot(x,y3,label=y3_label)
plt.plot(x,y4,label=y4_label)

plt.legend()
```
